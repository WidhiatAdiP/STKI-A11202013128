```python
phone = pd.read_csv('smartphone_cleaned_v5.csv')
print(phone)
```

```python
phone_filtered = phone[phone['processor_brand'] == "dimensity"]
print(phone_filtered)
```

```python
phone_brand = phone_filtered[phone_filtered['brand_name'] == "xiaomi"]
print(phone_brand)
```

```python
#Comparison Plot Dimensity and Snapdragon
dimensityOrsnapdragon = [np.count_nonzero(phone['processor_brand'] == "dimensity"), np.count_nonzero(phone['processor_brand'] == "snapdragon")]
label = ['dimensity', 'snapdragon']
plt.title('dimensity and snapdragon')
plt.pie(dimensityOrsnapdragon, labels = label, radius=1.2, startangle=60, autopct='%.1f%%', shadow=True)
plt.show()

#Comparison Plot Phone Based on Brand
phoneBrand = [np.count_nonzero(phone_filtered['brand_name'] == "xiaomi"), np.count_nonzero(phone['brand_name'] == "poco")]
label2 = ['xiaomi', 'poco']
plt.title('xiaomi and poco')
plt.pie(phoneBrand, labels = label2, radius=1.1, startangle=60, autopct='%.1f%%', shadow=True)
plt.show()
```
